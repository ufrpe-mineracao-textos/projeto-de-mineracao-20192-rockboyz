```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/projeto-de-mineracao-20192-rockboyz/questions.csv')

df
```

```
ans = df[df["Answer"]]
df["Awsner Text"] = pd.Series(np.diag(ans))


df["Correct"] = df["Question"].replace(r'_', df["Awsner Text"], regex=True)
list(df["Correct"])
```

```
df_clean = pd.DataFrame({'clean': df['Correct']})
sent = [row.split(' ') for row in df_clean['clean']]
sent
```

```
from gensim.models import Word2Vec
model = Word2Vec(sent, min_count=1,size= 50,workers=3, window =3, sg = 0)
```

```
print(model.predict_output_word("The newspaper editor received public opprobrium for publishing a graphic account of the politician's sexual harassment scandal; however, in the editor's defense, she knew the public would want to know every detail _ of the case.".split(' ')))
```
